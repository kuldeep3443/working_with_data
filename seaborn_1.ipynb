## Histogram with Seaborn

```python
sns.__version__
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
sns.displot(penguins, x="flipper_length_mm")
```

```python
sns.displot(penguins, x="bill_depth_mm")
```

```python
## Choosing the bin size

## binwidth is the width of a single bar

sns.displot(penguins, x="flipper_length_mm", binwidth=3)
```

```python
## Working with bins

sns.displot(penguins, x="flipper_length_mm", bins=20)
```

---

## Conditioning on other variables

```python
sns.displot(penguins, x="flipper_length_mm", hue="species")
```

```python
## By default, the different histograms are “layered” on top of each other and, in some cases, they may be difficult to 
  ## distinguish. One option is to change the visual representation of the histogram from a bar plot to a “step” plot:
    
sns.displot(penguins, x="flipper_length_mm", hue="species", element="step")    
```

```python
### Because displot() is a figure-level function and is drawn onto a FacetGrid, it is also possible to draw each individual 
 ## distribution in a separate subplot by assigning the second variable to col or row rather than (or in addition to) hue. 
    ## This represents the distribution of each subset well, but it makes it more difficult to draw direct comparisons:
    
sns.displot(penguins, x="flipper_length_mm", col="sex", multiple="dodge")    
```

## Plotting joint and marginal distributions

```python
sns.jointplot(data=penguins, x="bill_length_mm", y="bill_depth_mm")
```

```python
sns.jointplot(
    data=penguins,
    x="bill_length_mm", y="bill_depth_mm", hue="species",
    kind="kde"
)
```

```python
g = sns.JointGrid(data=penguins, x="bill_length_mm", y="bill_depth_mm")
g.plot_joint(sns.histplot)
g.plot_marginals(sns.boxplot)
```

---

## Plotting many distributions

```python
## The pairplot() function offers a similar blend of joint and marginal distributions. Rather than focusing on a 
 ## single relationship, however, pairplot() uses a “small-multiple” approach to visualize the univariate distribution of all 
    ## variables in a dataset along with all of their pairwise relationships:
    
sns.pairplot(penguins)    
```

```python
g = sns.PairGrid(penguins)
g.map_upper(sns.histplot)
g.map_lower(sns.kdeplot, fill=True)
g.map_diag(sns.histplot, kde=True)
```
